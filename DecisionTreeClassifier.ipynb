```python
import cv2 
import glob
import numpy as np
import pandas as pd
import random
random.seed(10)
from scipy import stats
import sklearn.metrics as metrics
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
```

```python
tr_img = []
for img in glob.glob("D:/computer science GIU/semester 6/Advanced Machine Learning/ass 2/Assignment 2 Dataset/Train/*.jpg"):
    n= cv2.imread(img)
    tr_img.append(n)
tr_img  
```

```python
te_img = []
for img in glob.glob("D:/computer science GIU/semester 6/Advanced Machine Learning/ass 2/Assignment 2 Dataset/Test/*.jpg"):
    n= cv2.imread(img)
    te_img.append(n) 
te_img    
```

```python
tr_text = open("D:/computer science GIU/semester 6/Advanced Machine Learning/ass 2/Assignment 2 Dataset/Train/Training Labels.txt", "r")
tr_data = tr_text.read().splitlines()
tr_data
```

```python
te_text = open("D:/computer science GIU/semester 6/Advanced Machine Learning/ass 2/Assignment 2 Dataset/Test/Test Labels.txt", "r")
te_data = te_text.read().splitlines()
te_data
```

```python
tr_1d = []
tr_arr = []
for i in range(2400):
    tr_arr = tr_img[i].flatten()
    tr_1d.append(tr_arr)
    tr_arr = []
tr_1d    
```

```python
te_1d = []
te_arr = []
for i in range(200):
    te_arr = te_img[i].flatten()
    te_1d.append(te_arr)
    te_arr = []  
te_1d    
```

```python
tr_df = pd.DataFrame(tr_1d)
```

```python
tr_df['Numbers'] = tr_data
```

```python
tr_df
```

```python
te_df = pd.DataFrame(te_1d)
```

```python
te_df['Numbers'] = te_data
```

```python
te_df
```

```python
sample = tr_df.sample(50)
```

```python
x_tr = sample.copy()
x_te = te_df.copy()
y_tr = sample['Numbers']
y_te = te_df['Numbers']
tree=[]
for i in range(10):
        x = DecisionTreeClassifier(random_state = 10, max_depth = 10)
        tree.append(x)
        x = x.fit(x_tr, y_tr)
tree        
```

```python
predictions = []
for i in range(10):
    predictions.append(tree[i].predict(x_te))
predictions
```

```python
mode=[]
for i in range(10):
    mode.append(stats.mode(predictions[i]))
mode    
```

```python
for i in range(10):    
    print(confusion_matrix(y_te, predictions[i]))
    print(metrics.accuracy_score(y_te, predictions[i]))
```
