```python
import cv2 
import glob
import numpy as np
import pandas as pd
import random
random.seed(10)
from scipy import stats
import sklearn.metrics as metrics
from warnings import simplefilter
simplefilter(action='ignore', category=FutureWarning)
```

```python
from sklearn.metrics import confusion_matrix
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
```

```python
#Read the Train folder of images
train_img = []
for img in glob.glob("D:/computer science GIU/semester 6/Advanced Machine Learning/ass 2/Assignment 2 Dataset/Train/*.jpg"):
    n= cv2.imread(img)
    train_img.append(n)
```

```python
#Convert the 2d image to 1d vector
train_array = []
train_all_arrays = []
for i in range(len(train_img)):
    train_array = train_img[i].flatten()
    train_all_arrays.append(train_array)
    train_array = []
```

```python
#Put the Train in dataframe
train_df = pd.DataFrame(train_all_arrays)
```

```python
#Read the text file
train_fileObject = open("D:/computer science GIU/semester 6/Advanced Machine Learning/ass 2/Assignment 2 Dataset/Train/Training Labels.txt", "r")
train_data = train_fileObject.read().splitlines()
```

```python
#Add the Label to the dataframe
train_df['Label'] = train_data
```

```python
#Read the Test folder of images
test_img = []
for img in glob.glob("D:/computer science GIU/semester 6/Advanced Machine Learning/ass 2/Assignment 2 Dataset/Test/*.jpg"):
    n= cv2.imread(img)
    test_img.append(n) 
```

```python
#Convert the 2d image to 1d vector
test_array = []
test_all_arrays = []
for i in range(len(test_img)):
    test_array = test_img[i].flatten()
    test_all_arrays.append(test_array)
    test_array = []     
```

```python
#Put the Test in dataframe
test_df = pd.DataFrame(test_all_arrays)
```

```python
#Read the text file
test_fileObject = open("D:/computer science GIU/semester 6/Advanced Machine Learning/ass 2/Assignment 2 Dataset/Test/Test Labels.txt", "r")
test_data = test_fileObject.read().splitlines()
```

```python
#Add the Label to the dataframe
test_df['Label'] = test_data
```

```python
#clf[2].predict_proba(x_test)
test_df
```

```python
def random_forest(number_of_sample, number_of_trees, maxDepth):
    A = train_df.sample(number_of_sample)
    x_train = A.copy()
    x_test = test_df.copy()
    y_train = A['Label']
    y_test = test_df['Label']
    clf=[]
    predictions = []
    mode=[]
    for i in range(number_of_trees):
        z = DecisionTreeClassifier(random_state = 10, max_depth = maxDepth)
        clf.append(z)
        z = z.fit(x_train, y_train)
    for i in range(number_of_trees):
        predictions.append(clf[i].predict(x_test))
    for i in range(number_of_trees):
        mode.append(stats.mode(predictions[i]))
    for i in range(number_of_trees):    
        print("Confusion:",confusion_matrix(y_test, predictions[i]))
        print("Accuracy:",metrics.accuracy_score(y_test, predictions[i]))    
    return mode    
random_forest(50, 10, 10)
```

```python
random_forest(50, 100, 10)
```

```python
random_forest(500, 100, 10)
```

```python
random_forest(1000, 100, 100)
```
