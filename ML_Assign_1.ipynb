```python

z=train.drop(["Species"],axis=1)
mat=z.to_numpy()
x_tilda = np.concatenate((mat, np.ones((105,1))), axis=1)
target1=[]
c=train['Species'].values
for i in range (105):
    if (c[i]=='Iris-versicolor'):
        target1+=[1]
    else:
        target1+=[-1]
w1 = np.linalg.inv(x_tilda.T @ x_tilda) @ x_tilda.T @ target1

w1
```
