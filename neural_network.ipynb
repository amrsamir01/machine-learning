```python
import tensorflow as tf
import cv2 
import glob
import numpy as np
import pandas as pd
from tensorflow import keras
from tensorflow.keras import layers
from PIL import Image
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
import os
os.environ['PYTHONHASHSEED'] = '0'
np.random.seed(37)
tf.random.set_seed(89)
```

```python
#Read the train folder
from skimage.io import imread_collection
train_img = np.array(imread_collection(r'D:/computer science GIU/semester 6/Advanced Machine Learning/ass 3/Assignment 3 Dataset/Assignment 3 Dataset/Train/*.jpg'))
```

```python
#Convert the 2d image to 1d vector
train_array = np.tile(0, (2400, 784))
for i in range(len(train_img)):
    train_array[i] = train_img[i].flatten()
```

```python
#Read the text file in the train folder
train_fileObject = open("D:/computer science GIU/semester 6/Advanced Machine Learning/ass 3/Assignment 3 Dataset/Assignment 3 Dataset/Train/Training Labels.txt", "r")
train_data = train_fileObject.read().splitlines()
train_data = [int(numeric_string) for numeric_string in train_data]
```

```python
#Read the Test folder of images
test_img = np.array(imread_collection(r'D:/computer science GIU/semester 6/Advanced Machine Learning/ass 3/Assignment 3 Dataset/Assignment 3 Dataset/Test/*.jpg'))
```

```python
#Convert the 2d image to 1d vector
test_array = np.tile(0, (200, 784))
for i in range(len(test_img)):
    test_array[i] = test_img[i].flatten()
```

```python
#Read the text file in the test folder
test_fileObject = open("D:/computer science GIU/semester 6/Advanced Machine Learning/ass 3/Assignment 3 Dataset/Assignment 3 Dataset/Test/Test Labels.txt", "r")
test_data = test_fileObject.read().splitlines()
test_data = [int(numeric_string) for numeric_string in test_data]
```

```python
#make new labels of 0 and 1 for the train data
train = []
for i in range(10):
    rows, cols = (240,10)
    arr = [[0]*cols]*rows
    arr[0][i] = 1
    for row in arr:
        train.append(row)
```

```python
#make new labels of 0 and 1 for the test data
test = []
for i in range(10):
    rows, cols = (20,10)
    arr = [[0]*cols]*rows
    arr[0][i] = 1
    for row in arr:
        test.append(row)
```

```python
#clasify the data
X_train = tf.constant(train_array, dtype=tf.float32)
Y_train = tf.constant(train, dtype=tf.float32)
X_test = tf.constant(test_array, dtype=tf.float32)
Y_test = tf.constant(test, dtype=tf.float32)
```

```python
#Standardize the data
scaler = StandardScaler()
scaled_X_train = scaler.fit_transform(X_train)
scaled_Y_train = scaler.fit_transform(Y_train)
scaled_X_test = scaler.fit_transform(X_test)
scaled_Y_test = scaler.fit_transform(Y_test)
```

```python
#one layer standardized neural network
def Standardize_oneLayer(num_of_neurons, learning_rate_i, Epochs):
    model = keras.Sequential()
    # Add our layers
    model.add(layers.Dense(
        num_of_neurons, # Amount of Neurons
        input_shape=(784,), # Define an input dimension because this is the first layer
        activation='relu' # Use relu activation function because all inputs are positive
    ))

    model.add(layers.Dense(
        10, # Amount of Neurons. We want one output
        activation='sigmoid' # Use sigmoid because we want to output a binary classification
    ))

    keras.optimizers.SGD(
        learning_rate = learning_rate_i
    )
    
    model.compile(
        loss='mean_squared_error', # The loss function that is being minimized
        optimizer='adam', # Our optimization function
        metrics=['accuracy'] # Metrics are different values that you want the model to track while training
    )
    
    model.fit(
        scaled_X_train,
        scaled_Y_train,
        epochs=Epochs,  
    )
    
    model.evaluate(
        scaled_X_test,
        scaled_Y_test
    )
    predict = model.predict(scaled_X_test)
    y_test_pred = [np.argmax(i) for i in predict]
    print(tf.math.confusion_matrix(labels = test_data, predictions = y_test_pred))
    print('Test Data accuracy: ', accuracy_score(test_data, y_test_pred)*100)
```

```python
Standardize_oneLayer(500, 0.001, 100)
```

```python
Standardize_oneLayer(500, 0.001, 200)
```

```python
#four layers standardized neural network
def Standardize_fourLayers_batch(num_of_neurons_1, num_of_neurons_2, num_of_neurons_3, num_of_neurons_4, learning_rate_i, Epochs):
    model = keras.Sequential()
    # Add our layers
    model.add(layers.Dense(
        num_of_neurons_1, # Amount of Neurons
        input_shape=(784,), # Define an input dimension because this is the first layer
        activation='relu' # Use relu activation function because all inputs are positive
    ))
    
    model.add(layers.Dense(
        num_of_neurons_2, # Amount of Neurons
        input_shape=(784,), # Define an input dimension because this is the first layer
        activation='relu' # Use relu activation function because all inputs are positive
    ))
    
    model.add(layers.Dense(
        num_of_neurons_3, # Amount of Neurons
        input_shape=(784,), # Define an input dimension because this is the first layer
        activation='relu' # Use relu activation function because all inputs are positive
    ))
    
    model.add(layers.Dense(
        num_of_neurons_4, # Amount of Neurons
        input_shape=(784,), # Define an input dimension because this is the first layer
        activation='relu' # Use relu activation function because all inputs are positive
    ))

    model.add(layers.Dense(
        10, # Amount of Neurons. We want one output
        activation='sigmoid' # Use sigmoid because we want to output a binary classification
    ))

    keras.optimizers.SGD(
        learning_rate = learning_rate_i
    )
    
    model.compile(
        loss='mean_squared_error', # The loss function that is being minimized
        optimizer='adam', # Our optimization function
        metrics=['accuracy'] # Metrics are different values that you want the model to track while training
    )
    
    model.fit(
        scaled_X_train,
        scaled_Y_train,
        epochs=Epochs,
        batch_size=1 
    )
    
    model.evaluate(
        scaled_X_test,
        scaled_Y_test
    )
    predict = model.predict(scaled_X_test)
    y_test_pred = [np.argmax(i) for i in predict]
    print(tf.math.confusion_matrix(labels = test_data, predictions = y_test_pred))
    print('Test Data accuracy: ', accuracy_score(test_data, y_test_pred)*100)
```
